#### The class distribution is imbalanced, with 0 being the majority class and 1 as the minority class.The dataset contains 500 instances labeled as 0 and 268 instances labeled as 1.
***To address the class imbalance issue, several techniques can be used such as Resampling Techniques(Oversampling/Undersampling), Class Weighting etc.***

**Oversampling:** 
Oversampling is a widely used technique to address class imbalance by increasing the number of instances in the minority class. There are various methods for oversampling, with **SMOTE (Synthetic Minority Over-sampling Technique)** being one of the most commonly used approaches.

```python
from imblearn.over_sampling import SMOTE

# Separate features and target variable
X = df.drop('Outcome', axis=1)  # Features
y = df['Outcome']  # Target variable

# Apply SMOTE to oversample the minority class
smote = SMOTE(random_state=42)
X_oversampled, y_oversampled = smote.fit_resample(X, y)

df_oversampled = pd.concat([pd.DataFrame(X_oversampled), pd.DataFrame(y_oversampled, columns=['Outcome'])], axis=1)
print(df_oversampled['Outcome'].value_counts())
df_oversampled.head()
```

To train and assess the models' performance, we'll employ the following methodology:

Utilizing GridSearchCV with Cross Validation for training and testing each model.This approach iterates through predefined hyperparameters, enabling us to fit models on the training set and fine-tune hyperparameters effectively.
GridSearchCV aids in hyperparameter tuning, providing optimal hyperparameters for each model.
This comprehensive evaluation strategy using a range of models, combined with GridSearchCV for hyperparameter tuning and Cross Validation, will enable us to select the most suitable model based on its predictive capabilities.

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import accuracy_score, average_precision_score, f1_score, confusion_matrix, classification_report, auc, roc_curve, roc_auc_score, precision_recall_curve

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_oversampled, y_oversampled, test_size=0.25, random_state=10)
```

### 2. Apply an appropriate classification algorithm to build a model. Compare various models with the results from KNN algorithm.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score

# Initialize classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machine': SVC(),
    'Naive Bayes': GaussianNB(),
    'AdaBoost': AdaBoostClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier()
}

# Train and evaluate each classifier
results = {}
for name, classifier in classifiers.items():
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    results[name] = {'Accuracy': accuracy, 'F1 Score': f1}

# Display results
print("Performance Comparison of Different Classification Models with KNN:")
print("-----------------------------------------------------------------")
print("{:<25} {:<15} {:<15}".format('Classifier', 'Accuracy', 'F1 Score'))
print("-----------------------------------------------------------------")
for name, metrics in results.items():
    print("{:<25} {:<15.4f} {:<15.4f}".format(name, metrics['Accuracy'], metrics['F1 Score']))
```

```python
models = []
model_accuracy = []
model_f1 = []
model_auc = []
```

## 1) Logistic Regression:

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression(max_iter=250)
logistic_regression.fit(X_train, y_train)

# Calculate accuracy scores for training and testing sets
train_accuracy = logistic_regression.score(X_train, y_train)
test_accuracy = logistic_regression.score(X_test, y_test)

print("Logistic Regression Model:")
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Testing Accuracy: {test_accuracy:.4f}")
```

**Performance evaluation and optimizing parameters using GridSearchCV:** 

```python
from sklearn.model_selection import GridSearchCV, cross_val_score

parameters = {'C': np.logspace(-5, 5, 50)}

# Create GridSearchCV with modified variable names
grid_search = GridSearchCV(estimator=LogisticRegression(max_iter=250), param_grid=parameters, cv=5, verbose=0)

# Fit GridSearchCV on the resampled data
grid_search.fit(X_oversampled, y_oversampled)

# Retrieve best parameters and best score from GridSearchCV
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Initialize a new Logistic Regression model with the best parameters
lr_best = LogisticRegression(C=best_params['C'], max_iter=250)

# Train the new Logistic Regression model on the training data
lr_best.fit(X_train, y_train)

# Calculate accuracy scores for the new model on training and testing sets
train_accuracy_best = lr_best.score(X_train, y_train)
test_accuracy_best = lr_best.score(X_test, y_test)

# Display the best parameters, best score, and accuracy scores
print("Best Parameters:", best_params)
print("Best Score:", best_score)
print("Logistic Regression Model with Best Parameters:")
print(f"Training Accuracy: {train_accuracy_best:.4f}")
print(f"Testing Accuracy: {test_accuracy_best:.4f}")
```

```python
# Predict probabilities for the positive outcome using the best Logistic Regression model
probs_best = lr_best.predict_proba(X_test)
probs_best = probs_best[:, 1]  # Keep probabilities for the positive outcome only

# Calculate AUC (Area Under the Curve) for the best model
auc_lr_best = roc_auc_score(y_test, probs_best)
print('AUC for Best Model: %.3f' % auc_lr_best)

# Calculate ROC curve for the best model
fpr_best, tpr_best, thresholds_best = roc_curve(y_test, probs_best)

# Plot ROC curve for the best model
plt.plot([0, 1], [0, 1], linestyle='--', label='No Skill')  # Plot no skill line
plt.plot(fpr_best, tpr_best, marker='.', label='Logistic Regression')  # Plot ROC curve for the model
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for Best Logistic Regression Model")
plt.legend()
plt.show()
```

```python
# Predict class values for the best Logistic Regression model
pred_y_test_best = lr_best.predict(X_test)

# Calculate precision-recall curve for the best model
precision_best, recall_best, thresholds_best = precision_recall_curve(y_test, probs_best)

# Calculate F1 score for the best model
f1_best = f1_score(y_test, pred_y_test_best)

# Calculate precision-recall AUC for the best model
auc_lr_pr_best = auc(recall_best, precision_best)

# Calculate average precision score for the best model
ap_best = average_precision_score(y_test, probs_best)

# Display F1 score, precision-recall AUC, and average precision score for the best model
print('F1 Score: %.3f' % f1_best)
print('Precision-Recall AUC: %.3f' % auc_lr_pr_best)
print('Average Precision Score: %.3f' % ap_best)

# Plot Precision-Recall curve for the best model
plt.plot([0, 1], [0.5, 0.5], linestyle='--', label='No Skill')  # Plot no skill line
plt.plot(recall_best, precision_best, marker='.', label='Logistic Regression')  # Plot precision-recall curve
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve for Best Logistic Regression Model")
plt.legend()
plt.show()
```

```python
models.append('LR')
model_accuracy.append(accuracy_score(y_test, pred_y_test_best))
model_f1.append(f1_best)
model_auc.append(auc_lr_pr_best)
```
